```python
import pandas as pd
```

```python
df = pd.read_csv('arrests.csv')
#if the data is in an excel file
df_xlsx = pd.read_excel('arrests.xlsx')
#if the data is in a text format,specify a delimiter(something which is separating the data)
df = pd.read_csv('arrests.txt',delimiter='\t')
```

```python
print(df.head(5))
#bottom rows
print(df.tail(5))
```

Read the data nicely in pandas

```python
#read headers
print(df.columns)
#read each column
print(df['citizen'][0:3])
#or
print(df.citizen)
#if we want to access different columns of data
print(df[['citizen','colour','age']])
#read each row
print(df.head(3))
#read a specifically indexed row
print(df.iloc[2:6])
#read a specific location(row,column)
print(df.iloc[2,4])
#find specific data
df =df.loc[(df['released']=='Yes') or (df['age']==17 )]
```

```python
#iterate over a row
for index,row in df.iterrows():
 print(index,row['colour'])
```

```python
df.describe()
```

```python
df.sort_values('year',ascending = False)
```

Making changes to the data

```python
#adding a column
df['Year']= df['year']+df['age']
df.head(5)
#or
df['Total']=df.iloc[:,3:5].sum(axis=1)
#shift a column
cols = list(df.columns)
df=df[cols[0:4]+cols[-1]+cols[4:9]]
#deleting a column
df = df.drop(columns=['Year'])
df.head(5)
```

```python
#save the new file
df.to_csv('modified.csv',index=False)
```

Aggregate Statistics(groupby)

```python
df.groupby(['sex']).mean().sort_values('year',ascending = False)
```

```python
df.groupby(['age']).sum()
#only want to add a new column named count and access it for convenience
df['count']=1 
df.groupby(['age']).count()['count']
#can also groupby multiple parameters
df.groupby(['age','sex']).count()['count']
```
